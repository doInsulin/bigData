```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movie.csv")
df.head()
```

```python
df.dtypes
```

```python
df1=df.dtypes
df1=pd.DataFrame(df1)
df1.columns=['types']
df1[(df1['types']=='object')]
```

```python
df[(df['title_year']>2000)
   &(df['country']=='USA')
   &(df['language']=='English')
   &(df['genres'].str.contains('Action'))]
```

```python
df1.loc[df1['types'] == 'object'] = df1.loc[df1['types'] == 'object'].applymap(lambda x: np.nan)
```

```python

# 首先，筛选内容分级为'PG-13'的电影。
df_pg_13 = df[df['content_rating'] == 'PG-13']

# 然后，按照年份分组，并对num_voted_users和num_user_for_reviews进行汇总。
grouped_by_year = df_pg_13.groupby('title_year').agg({
    'num_voted_users': 'sum',
    'num_user_for_reviews': 'sum'
}).reset_index()

# 使用print输出按年份分组的结果
print("Grouped by Year:")
print(grouped_by_year)

# 接下来，按照题材分组，并对num_voted_users和num_user_for_reviews进行汇总。
grouped_by_genre = df_pg_13.groupby('genres').agg({
    'num_voted_users': 'sum',
    'num_user_for_reviews': 'sum'
}).reset_index()

# 使用print输出按题材分组的结果
print("\nGrouped by Genre:")
print(grouped_by_genre)
```
